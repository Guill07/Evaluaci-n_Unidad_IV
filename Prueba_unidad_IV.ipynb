# üßÆ Ejercicio 1: Suma de D√≠gitos (Valor - 5 pts)

### Descripci√≥n
Escribe un programa que calcule la suma de los d√≠gitos de un n√∫mero entero positivo ingresado por el usuario. El programa debe verificar que el n√∫mero sea positivo utilizando estructuras condicionales. Para el c√°lculo, debes utilizar un ciclo **while** que extraiga cada d√≠gito del n√∫mero mediante operaciones de m√≥dulo y divisi√≥n entera.

---

### ‚úÖ Salida esperada

- Entrada: `numero = 1234`

```python
# Aqu√≠ se realiza el ejercicio 1
numero_del_usuario= int(input("Ingrese un numero entero positivo"))
suma_digitos= 0
numero_calcular= numero_del_usuario
if numero_del_usuario>0:
    numero_calcular= numero_del_usuario
    while numero_calcular>0:
        digito= numero_calcular % 10
        suma_digitos= suma_digitos + digito
        numero_calcular= numero_calcular // 10 
    print("la suma de los digitos es:", suma_digitos)
    
else:
    ("error")


```

# üî§ Ejercicio 2: Contador de Vocales y Consonantes (Valor - 5 pts)

### Descripci√≥n
Escribe un programa que cuente la cantidad de vocales y consonantes en una frase ingresada por el usuario. El programa debe ignorar espacios, n√∫meros, s√≠mbolos y cualquier car√°cter que no sea una letra del alfabeto. Debes considerar tanto may√∫sculas como min√∫sculas como equivalentes (case-insensitive).

Utiliza un ciclo **for** para iterar sobre cada car√°cter de la frase y estructuras condicionales para clasificar las letras en vocales o consonantes.

---

### ‚úÖ Salida esperada
El programa debe devolver el conteo de vocales y consonantes por separado.

**Ejemplo:**
- Entrada: `frase = "Hola Mundo!"`  
- Salida: `Vocales: 4, Consonantes: 5`  

```python
# Aqu√≠ se realiza el ejercicio 2
frase_del_usuario= input("ingrese una frase para contar vocales y consonantes")
cuenta_vocales=0
cuenta_consonantes=0

vocales="aeiou"
frases_minusculas= frase_del_usuario.lower()
for caracter in frases_minusculas:
    if caracter.isalpha():
        if caracter in vocales:
            cuenta_vocales=cuenta_vocales+1
        else:
            cuenta_consonantes= cuenta_consonantes + 1
print("resultado del conteo")
print(f"frase ingresada: '{frase_del_usuario}'")
print(f"vocales:{cuenta_vocales}")
print(f"consonantes:{cuenta_consonantes}")
```

# üìê Ejercicio 4: M√°ximo Com√∫n Divisor (MCD) (Valor - 5 pts)

### Descripci√≥n
Escribe un programa que calcule el m√°ximo com√∫n divisor (MCD) de dos n√∫meros enteros positivos ingresados por el usuario. Debes implementar el algoritmo de Euclides, que utiliza divisiones sucesivas para encontrar el MCD.

El algoritmo funciona de la siguiente manera:
- Mientras el segundo n√∫mero no sea cero:
    - El primer n√∫mero se reemplaza por el segundo n√∫mero.
    - El segundo n√∫mero se reemplaza por el residuo de la divisi√≥n del primer n√∫mero entre el segundo.
- El MCD es el valor final del primer n√∫mero cuando el segundo n√∫mero sea cero.

Utiliza un ciclo while para implementar este algoritmo.

---

### ‚úÖ Salida esperada
El programa debe devolver el MCD de los dos n√∫meros ingresados.

**Ejemplo:**
- Entrada: `a = 48, b = 18`  
- Salida: `El MCD es: 6`  

```python
# Aqui se resuleve el ejercicio 4
#Le pedimos al usuario que
N_A = int(input("Ingrese el primer n√∫mero (A): "))
N_B = int(input("Ingrese el segundo n√∫mero (B): "))

#Apliccamos el while utilizando el algoritmo de Euclides    
while N_B != 0:
        residuo = N_A % N_B
        N_A = N_B
        N_B = residuo
        
mcd_resultado = N_A

print(f"El MCD es: {mcd_resultado}")
```
